```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn
import scipy.stats

seaborn.set()
table_file = r'D:\FBB\bioinf\4 семестр\inf_ht\pandas\Automobile_data.csv' # input_table path

data = pd.read_csv(table_file)
```

```python
# Horsepower

data.iloc[[130, 131]] # cars with unknown horsepower 
```

```python
data_no_unknown_hp = data.drop([130, 131])
data_no_unknown_hp['horsepower'] = data_no_unknown_hp['horsepower'].astype('int64')
group_brand_hp = data_no_unknown_hp.groupby('brand')
```

```python
hp_mean = group_brand_hp.horsepower.mean()
hp_sd = group_brand_hp.horsepower.std()
hp_mean_df = pd.DataFrame({'Brand':hp_mean.index, 'Mean Horsepower':hp_mean.values})
hp_mean_df = hp_mean_df.sort_values('Mean Horsepower', ascending=True)
hp_mean_df.index = range(21)
hp_mean_df
```

```python
plt.figure(figsize=(11, 11))
plt.bar(x = 'Brand', height = 'Mean Horsepower', data = hp_mean_df, color = 'r')
plt.xticks(rotation=90)
plt.legend(title = 'Mean horsepower for different car brands', title_fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(11, 11))
x = 'brand'
y = 'horsepower'
plt.scatter(x, y, data=data_no_unknown_hp, linewidth=0.4)
plt.xticks(rotation=90)
plt.legend(title = 'Horsepower dispersion for different car brands', title_fontsize=15)
plt.show()
```

```python
# Prices

data.iloc[[9, 44, 45, 129]] # cars with unknown price
```

```python
data_no_unknown_prices = data.drop([9, 44, 45, 129])
data_no_unknown_prices['price'] = data_no_unknown_prices['price'].astype('int64')
group_brand_price = data_no_unknown_prices.groupby('brand')
```

```python
price_mean = group_brand_price.price.mean()
price_sd = group_brand_price.price.std()
price_mean_df = pd.DataFrame({'Brand':price_mean.index, 'Mean Price':price_mean.values})
price_mean_df = price_mean_df.sort_values('Mean Price', ascending=True)
price_mean_df.index = range(22)
price_mean_df
```

```python
plt.figure(figsize=(11, 11))
plt.bar(x = 'Brand', height = 'Mean Price', data = price_mean_df, color = 'r')
plt.xticks(rotation=90)
plt.legend(title = 'Mean prices for different car brands', title_fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(11, 11))
x = 'brand'
y = 'price'
plt.scatter(x, y, data=data_no_unknown_prices, linewidth=0.4)
plt.xticks(rotation=90)
plt.legend(title = 'Price dispersion for different car brands', title_fontsize=15)
plt.show()
```

```python
over_mean = data_no_unknown_hp[data_no_unknown_hp['horsepower'] > hp_mean_df['Mean Horsepower'].mean()][['brand', 'horsepower', 'price']].drop([9,129])
over_mean.sort_values('price', ascending=True)
over_mean.groupby('price')
over_mean.index = range(57)
over_mean['price'] = over_mean['price'].astype('int64')
over_mean
```

```python
over_df = over_mean.shape[0] - 2
over_sxy = ((((over_mean['horsepower']-over_mean['horsepower'].mean())) * (over_mean['price']-over_mean['price'].mean())).sum())/(over_df+1)
over_sxx = (((over_mean['horsepower']-over_mean['horsepower'].mean())**2).sum())/(over_df+1)
over_syy = (((over_mean['price']-over_mean['price'].mean())**2).sum())/(over_df+1)
over_r = over_sxy/((over_syy*over_sxx)**0.5)
over_t = over_r/((over_df/(1-over_r**2))**0.5)
over_p_value = t.pdf(over_t, over_df)

print('H0: между столбцами \'price\' b \'horsepower\' существует линейная зависимость')
print('Поскольку p-value = '+ str(round(over_p_value, 3))+ ', то на 5% уровне значимости мы принимаем H0')
```
